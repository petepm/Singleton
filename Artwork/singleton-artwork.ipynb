```python
from matplotlib import font_manager
x = np.linspace(0, 6, num=100)
y = np.exp(x)
phi = np.linspace(0, 2*np.pi, num=100)
y2 = 2 * 150 * (1 - np.cos(phi)) * np.cos(phi) + 300 # + 325
x2 = 2 * 1 * (1 - np.cos(phi)) * np.sin(phi) + 3
fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(x, y, lw=8, label="$H$")
ax.plot(x, y, lw=8, label=r"$\frac{\partial H}{\partial t}$", linestyle="dotted", color="k")
ax.plot(x2, y2, lw=10, color="r", alpha=0.8, label="\u2764")

font = font_manager.FontProperties(family='sans-serif',
                                   weight='bold',
                                   style='normal', size=16)
ax.legend(prop=font)
ax.set_xlabel("Pete Peterson")
ax.set_ylabel("Singleton")
fig.tight_layout()
fig.savefig("singleton-artwork-raw.jpg", dpi=300)
```
