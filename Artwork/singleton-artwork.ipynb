```python
from matplotlib import font_manager
x = np.linspace(0, 6, num=100)
y = np.exp(x)
phi = np.linspace(0, 2*np.pi, num=100)
y2 = 2 * 80 * (1 - np.cos(phi)) * np.cos(phi) + 325
x2 = 2 * 1 * (1 - np.cos(phi)) * np.sin(phi) + 3
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y, lw=6, label="$H$")
ax.plot(x, y, lw=6, label=r"$\frac{\partial H}{\partial t}$", linestyle="dotted")
ax.plot(x2, y2, lw=6, color="r", alpha=0.25, label="\u2764")

font = font_manager.FontProperties(family='sans-serif',
                                   weight='bold',
                                   style='normal', size=16)
ax.legend(prop=font)
ax.set_xlabel("Pete Peterson")
ax.set_ylabel("Singleton")
# ax.legend()
```
